### <a name="mark_14"></a>Utilización De La "api.extensions" De Pandas.
[Index](#index)

Tienes razón, puede ser confuso entender los beneficios de usar `@pd.api.extensions.register_dataframe_accessor` en comparación con simplemente crear una clase y llamar a los métodos de esa clase directamente en un DataFrame de Pandas.

El uso de `@pd.api.extensions.register_dataframe_accessor` y la creación de un accessor personalizado tienen algunos beneficios potenciales:

1. **Organización y legibilidad del código**: Al registrar un accessor personalizado, puedes agrupar métodos y atributos relacionados en una clase dedicada, lo que facilita la comprensión y el mantenimiento del código. Esto puede ser especialmente útil cuando tienes muchas funcionalidades adicionales que deseas agregar a los DataFrames y quieres mantener el código organizado y estructurado.

2. **Acceso más intuitivo**: Al registrar un accessor personalizado, puedes acceder a los métodos y atributos adicionales directamente desde un objeto DataFrame de Pandas utilizando una sintaxis más intuitiva. En lugar de llamar a la clase y pasar el DataFrame como argumento, puedes acceder a los métodos de la clase a través de la notación de punto directamente en el DataFrame. Esto puede hacer que el código sea más legible y natural.

3. **Extensibilidad y reutilización**: Al crear un accessor personalizado, puedes agregar funcionalidades adicionales a los DataFrames de Pandas que no están disponibles de forma predeterminada. Esto te permite extender la funcionalidad de Pandas según tus necesidades específicas. Además, si necesitas utilizar estas funcionalidades en múltiples proyectos o scripts, puedes reutilizar el accessor personalizado registrándolo en cada proyecto sin tener que copiar y pegar el código en múltiples lugares.

4. **Compatibilidad con el ecosistema de Pandas**: Al utilizar el mecanismo de accessor personalizado, te beneficias de la compatibilidad con el ecosistema de Pandas. Esto significa que puedes aprovechar otras funcionalidades y herramientas de Pandas que también reconocen y trabajan con accessors personalizados. Por ejemplo, algunas librerías y extensiones de Pandas pueden interactuar directamente con los accessors registrados, lo que puede facilitar el uso y la integración de tu código en el ecosistema de Pandas.

En resumen, si bien puedes lograr funcionalidades similares creando una clase y llamando a sus métodos directamente en un DataFrame, el uso de `@pd.api.extensions.register_dataframe_accessor` ofrece beneficios en términos de organización, legibilidad, acceso intuitivo y compatibilidad con el ecosistema de Pandas. La elección de utilizar un accessor personalizado o no dependerá de tus necesidades y preferencias en el desarrollo de tu proyecto.

El mensaje "<pandas.core.strings.accessor.StringMethods at 0x7f40428f6390>" que se muestra al ejecutar `df.a.str` indica que estás accediendo al accessor `StringMethods` de Pandas para manipular las cadenas de texto en la columna "a" del DataFrame `df`.

El accessor `StringMethods` proporciona una serie de métodos para realizar operaciones específicas en cadenas de texto dentro de un DataFrame de Pandas. Al acceder a `df.a.str`, obtienes una instancia del accessor `StringMethods` asociado a la columna "a" del DataFrame.

A partir de este punto, puedes usar los métodos proporcionados por `StringMethods` para realizar diversas operaciones en las cadenas de texto de la columna "a". Algunos ejemplos de los métodos disponibles incluyen `upper()`, `lower()`, `replace()`, `contains()`, `split()`, entre otros. Puedes encadenar estos métodos para realizar transformaciones o consultas más complejas en las cadenas de texto.

Aquí tienes un ejemplo de cómo podrías usar el accessor `StringMethods` para convertir todas las cadenas de la columna "a" en mayúsculas:

```python
df.a.str.upper()
```

En este caso, `df.a.str.upper()` aplicaría el método `upper()` del accessor `StringMethods` a cada elemento de la columna "a", lo que resultaría en todas las cadenas convertidas a mayúsculas.

En resumen, al ejecutar `df.a.str`, el mensaje "<pandas.core.strings.accessor.StringMethods at 0x7f40428f6390>" indica que estás accediendo al accessor `StringMethods` para realizar operaciones específicas en las cadenas de texto de la columna "a" del DataFrame `df`.

### <a name="mark_15"></a>Creando La Conexión Con "api.extensions" De Pandas.
[Index](#index)

````python
@pd.api.extensions.register_dataframe_accessor("missing_values")

class MissingMethods:
    def __init__(self, pandas_obj):
        self._df = pandas_obj
    
    def count_nulls(self):
        return self._df.isna().sum().sum()
    
    def count_complete_01(self):
        return self._df.notna().sum().sum()
    
    #otra forma de count_complete_01
    def count_complete_02(self):
        return self._df.size - self._df.missing_values.count_nulls()
````

### <a name="mark_16"></a>Análisis De Valores Faltantes.
[Index](#index)

Nota Importante:

La diferencia entre las dos líneas de código es cómo se accede a los datos cargados desde un archivo R en formato RDA utilizando la biblioteca `pyreadr`.

En el caso a):

```python
riskfactors_df = pyreadr.read_r("./data/riskfactors.rda")
```

Aquí, `pyreadr.read_r("./data/riskfactors.rda")` carga los datos desde el archivo RDA y devuelve un objeto de tipo `pyreadr.ReadRDS`. Este objeto contiene los datos cargados, pero aún no se ha accedido a ellos. En esta línea de código, se asigna el objeto completo a la variable `riskfactors_df`.

En el caso b):

```python
riskfactors_df = pyreadr.read_r("./data/riskfactors.rda").get("riskfactors")
```

Aquí, `pyreadr.read_r("./data/riskfactors.rda")` carga los datos desde el archivo RDA y devuelve un objeto de tipo `pyreadr.ReadRDS`. Luego, se utiliza el método `get("riskfactors")` en este objeto para acceder a un dataframe específico dentro del archivo RDA. En esta línea de código, solo se asigna el dataframe llamado "riskfactors" a la variable `riskfactors_df`.

En resumen, la diferencia entre las dos líneas de código es que en la primera línea se asigna el objeto completo devuelto por `pyreadr.read_r` a la variable `riskfactors_df`, mientras que en la segunda línea se utiliza el método `get` para acceder específicamente al dataframe llamado "riskfactors" dentro del objeto y se asigna solo ese dataframe a la variable `riskfactors_df`, en el cual se puede utilizar diferentes métodos como "isna()" u otros. Dependiendo de tus necesidades, puedes elegir una u otra forma de acceso a los datos cargados desde el archivo RDA.

### <a name="mark_28"></a>Tipos de Valores Faltantes, MCAR, MAR, MNAR.
[Index](#index)

Tipos de Valores Faltantes
Tenemos tres tipos de mecanismo de valores faltantes:

Missing Completely at Random (Perdidos completamente al azar)
Missing at Random (Perdidos al azar)
Missing not at Random (Perdidos no al azar)
Missing Completely at Random (MCAR)
Son datos ocasionados por herramientas que dejan de funcionar sin ninguna razon detras. La localizacion de los valores faltantes ocurre completamente al azar, esto no depende de ningun otro dato

Missing at Random (MAR)
Las herramientas necesitan mantenimiento periodico para asegurar su funcionamiento constante. En las fechas donde se le hace mantenimiento sabemos que ocurrira una falta de datos por lo que esos datos faltantes son localizados. La localizacion de los valores faltantes en el conjunto de datos depende de otros valores observados

Missing not at Random (MNAR)
Las herramientas tienen limites. Al tratar de hacer seguimientos fuera de su rango de medicion, se generan valores faltantes. La localizacion de los valores faltantes en el conjunto de datos dependen de los valores faltantes en si mismos

Puedo tener seguridad sobre que mecanismo de valores faltantes es correcto para mis datos?

No, no podemos tener seuridad sobre cual mecanismo incidio sobre los datos, pero a traves de analisis y conocimiento del tema puedes hacer suposiciones razonables

### <a name="mark_30"></a>Visualización De Valores Faltantes De Dos Variables.
[Index](#index)

````python
#Definimos una función dummy que agrega parámetros random, para poder visualizarlos en un eje.
def column_fill_with_dummies(
    column: pd.Series,
    proportion_below: float=0.10,#para separar valores faltantes de valores comunes.
    jitter: float=0.075,         #para separa valores solapados.
    seed: int=42                 #componente aleatoreo, 42 por default.
) -> pd.Series:
    
    column = column.copy(deep=True)
    
    #Extract metadata values
    missing_mask=column.isna()
    number_missing_values=missing_mask.sum()
    column_range=column.max()-column.min()
    
    #Shift data
    column_shift=column.min()-column.min()*proportion_below
    
    #Create the "jitter" (noise) to be added around the points.
    np.random.seed(seed)
    column_jitter=(np.random.rand(number_missing_values)-2)*column_range*jitter
    
    #Save new dummy data.
    column[missing_mask]=column_shift+column_jitter
    
    return column
    
````

````python
(
    riskfactors_df
    .select_dtypes(
        exclude="category"
    )
    .pipe(
        lambda df: df[df.columns[df.isna().any()]]
    )
    .missing.bind_shadow_matrix(true_string=True, false_string=False)
    .apply(
        lambda column: column if "_NA" in column.name else column_fill_with_dummies(column, proportion_below=0.05, jitter=0.075)
    )
    .assign(
        nullity=lambda df: df.weight_lbs_NA | df.height_inch_NA
    )
    .pipe(
        lambda df:(
        sns.scatterplot(
            data=df,
            x="weight_lbs",
            y="height_inch",
            hue="nullity"
        )
        )
    )
)
````

Existen valores faltantes relacionados con otras variables en el conjunto de datos? 

La correlacion de nulidad nos ayuda a encontrar relaciones entre las variables en funcion de su nulidad.

- (Correlación = 1): una variable n = Nulo --> otra variable x = Nulo 
- (Correlación = -1): una variable n = Nulo --> otra variable x = No Nulo 
- (Correlación = <1): Cercano al 100% de una variable n = Nulo --> otra variable x = Nulo
- (Correlación = >1): Cercano al 100% de una variable n = Nulo --> otra variable x = No Nulo

### <a name="mark_32"></a>Eliminación de valores faltantes.
[Index](#index)

Eliminacion de Valores Faltantes
Cuando se trata de valores faltantes tenemos dos aproximaciones para el tratamiento de los mismos:

Eliminacion de valores faltantes: La eliminación de valores faltantes asume que los valores faltantes están perdidos completamente al azar (MCAR). En cualquier otro caso, realizar una eliminación de valores faltantes podrá ocasionar sesgos en los análisis y modelos subsecuentes. Por tanto, es importante investigar que mecanismos tienen los datos para asi evitar los sesgos a la hora de eliminarlos
Imputacion de valores faltantes: consiste en añadir valores en los datos faltantes para mantener el registro en el dataset
